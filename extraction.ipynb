```python
import os
import librosa
import numpy as np
import pandas as pd

# 1. Le chemin vers ton dossier 'data' (qui contient Actor_01, Actor_02...)
dataset_path = 'data/'

# Ces listes vont stocker nos résultats temporairement
features = []
labels = []

print("⏳ Début de l'extraction des données... Cela peut prendre une ou deux minutes.")

# 2. Parcourir tous les dossiers des acteurs (Actor_01, Actor_02, etc.)
for actor_dir in os.listdir(dataset_path):
    actor_path = os.path.join(dataset_path, actor_dir)
    
    # On vérifie que c'est bien un dossier
    if os.path.isdir(actor_path):
        
        # 3. Parcourir chaque fichier audio .wav à l'intérieur
        for file_name in os.listdir(actor_path):
            if file_name.endswith('.wav'):
                file_path = os.path.join(actor_path, file_name)
                
                try:
                    # ÉTAPE A : Trouver l'émotion
                    # Le nom est sous ce format : 03-01-05-01-01-01-01.wav
                    # On coupe aux tirets et on prend le 3ème chiffre (qui représente l'émotion)
                    emotion = int(file_name.split('-')[2])
                    
                    # ÉTAPE B : Lire l'audio
                    # librosa transforme le son en un tableau de nombres (y)
                    y, sr = librosa.load(file_path, sr=None)
                    
                    # ÉTAPE C : Extraire les 40 coefficients MFCC
                    mfccs = librosa.feature.mfcc(y=y, sr=sr, n_mfcc=40)
                    
                    # ÉTAPE D : Faire la moyenne pour avoir la même taille pour tous les fichiers
                    mfccs_mean = np.mean(mfccs.T, axis=0)
                    
                    # On sauvegarde les résultats dans nos listes
                    features.append(mfccs_mean)
                    labels.append(emotion)
                    
                except Exception as e:
                    print(f"⚠️ Erreur de lecture sur le fichier {file_name}: {e}")

# 4. Transformer nos listes en un beau tableau structuré (DataFrame)
df = pd.DataFrame(features)
df['emotion'] = labels

# 5. Sauvegarder ce tableau dans un fichier CSV définitif
df.to_csv('ravdess_features.csv', index=False)

print("✅ Extraction terminée avec succès !")
print(f"📊 Nombre de fichiers analysés : {df.shape[0]}")
print("📁 Le fichier 'ravdess_features.csv' a bien été créé dans ton dossier.")
```
